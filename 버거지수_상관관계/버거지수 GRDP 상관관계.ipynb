```python
import pandas as pd
pd.set_option('mode.chained_assignment',  None) # Warning 방지용
import numpy as np

ratio = pd.read_csv('./버거지수.csv', index_col=0, encoding='CP949', engine='python')

ratio.head() #작업내용 확인용 출력
```

```python
ratio = ratio.dropna()
```

```python
ratio = ratio.drop(['x','y'], axis=1)
ratio.head()
```

```python
import matplotlib.pyplot as plt

X = ratio.B_ratio.values
Y = ratio.GP.values
plt.scatter(X,Y, alpha=0.7)
plt.title('burger - GRDP/P')
plt.xlabel('burger')
plt.ylabel('GRDP/P')
plt.show()
```

```python
X = ratio.B_ratio.values
Y = ratio.인구수.values
plt.scatter(X,Y, alpha=0.7)
plt.title('burger - population')
plt.xlabel('burger')
plt.ylabel('population')
plt.show()
```

```python
X = ratio.B_ratio.values
Y = ratio.GRDP.values
plt.scatter(X,Y, alpha=0.7)
plt.title('burger - GRDP')
plt.xlabel('burger')
plt.ylabel('GRDP')
plt.show()
```

```python
import seaborn as sns

ht = sns.heatmap(ratio.corr(),annot=True, square = True)
```

```python
ht.get_figure().savefig('heatmap.png')
```
